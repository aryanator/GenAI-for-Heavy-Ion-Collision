# Data Prep

```python
#This function reads in the dipole from file 'file' at x_Bjorken 'xbj'
#It uses Heikki's dipole amplitude code to extract the r dependent dipole
#at xbj doing the required interpolations
def read_dipole(file, xbj):
    try:
        if os.path.exists(file):  # Check if the file exists
            result = dipole_module.dipole_amplitude(file, xbj)
            lines = result.strip().split('\n')
        
            # Initialize c2_value
            c2_value = None

            # Read C2 from first line
            if lines and lines[0].startswith("#"):
                c2_value = float(lines[0].split()[1])

            # Remove the first line
            if lines:
                lines = lines[1:]
        
            data = [list(map(float, line.split())) for line in lines]
            result_array = np.array(data).T
        
            return result_array, c2_value
        else:
            print(f"File {file} does not exist.")
            return None, None  # Return None values if the file doesn't exist
    except Exception as e:
        print(f"An error occurred while processing {file} with xbj={xbj}: {e}")
        return None, None  # Return None values in case of an error

    
# Folder containing data files
data_folder = "/kaggle/input/full-data/training_fix"

# Dynamically list all files in the folder
file_names = [os.path.join(data_folder, file) for file in os.listdir(data_folder) if os.path.isfile(os.path.join(data_folder, file))]

# Number of files in the folder
max_files = len(file_names)
```

```python
#Load data

#array of x values
exponent_values = np.linspace(-6, -2, 20 )
x_values = 10.**(exponent_values)

#Initialize initial data array
initial_data = []

xdim = len(x_values)
print("Number of x values =",xdim)

target_data = []
  
max_files = len(file_names)#5000

for i in range(xdim):
    #Set target x value
    xbj_target = x_values[i]
    target_data_aux = []
    count = 0
    for file in tqdm(file_names, desc="Loading training and validation data", unit="file"):
        #print(file)
        #print("x value =",xbj_target)
        result_array, c2_value = read_dipole(file, xbj_target)
        if i == xdim-1:
            # Append c2_value to result_array[1]
            target_data_aux.append(result_array[1])
            new_result_array = np.append(result_array[1], c2_value)
            initial_data.append(new_result_array)
        else:
            target_data_aux.append(result_array[1])
        count = count + 1
        if count >= max_files:
            break
    target_data.append(np.array(target_data_aux))
   
            
input_values = np.vstack(initial_data)
r = result_array[0]
print(len(target_data))
```

Modify data to include x_bj as parameter

```python
# Initialize the list to hold modified arrays
modified_data = []

for i in range(xdim):
    # Set target x value
    xbj_target = x_values[i]

    # Append xbj_target to each array in input_values
    appended_input_values = np.array([np.append(arr, xbj_target) for arr in input_values])

    # Store the modified arrays
    modified_data.append(appended_input_values)

# Convert the list of arrays to a single numpy array (if needed)
modified_data = np.array(modified_data)

print(modified_data.shape)  # This will show the shape of the modified data array

input_values_modified= [item for sublist in modified_data for item in sublist]
```

```python
target_values_modified = []
for sub_array in target_data:
    target_values_modified.extend(sub_array)

print(len(target_values_modified))
print(len(input_values_modified))
```

# LOAD MODELS

```python
# Load the scaler model
with gzip.open('/kaggle/input/rf/scikitlearn/default/1/scaler_v1_1.pkl', 'rb') as f:
    scaler = pickle.load(f)

# Load the Random Forest model
with gzip.open('/kaggle/input/rf/scikitlearn/default/1/rf_v1_1.pkl', 'rb') as f:
    rf_model_small = pickle.load(f)

print("Scaler and Random Forest model loaded successfully!")
```

# TRAIN THE MODEL ON NEW DATA

```python
# Assume new_input_values and new_target_values are your new data arrays
new_input_values_modified = np.array(input_values_modified)

# Split the data into training and validation sets (if needed)
X_train_new, X_val_new, y_train_new, y_val_new = train_test_split(new_input_values_modified, target_values_modified, test_size=0.2, random_state=42, shuffle=True)

# Separate the features to be standardized and the feature to be kept raw
X_train_new_features = X_train_new[:, :-1]
X_train_new_last_feature = X_train_new[:, -1].reshape(-1, 1)

X_val_new_features = X_val_new[:, :-1]
X_val_new_last_feature = X_val_new[:, -1].reshape(-1, 1)

# Standardize the features using the loaded scaler
X_train_new_features_standardized = scaler.transform(X_train_new_features)
X_val_new_features_standardized = scaler.transform(X_val_new_features)

# Concatenate the standardized features with the raw last feature
X_train_new_standardized = np.concatenate([X_train_new_features_standardized, X_train_new_last_feature], axis=1)
X_val_new_standardized = np.concatenate([X_val_new_features_standardized, X_val_new_last_feature], axis=1)
```

```python
# Retrain the Random Forest model on the new standardized data
rf_model_small.fit(X_train_new_standardized, y_train_new)

# Make predictions on the new validation set
y_val_new_pred_rf = rf_model_small.predict(X_val_new_standardized)

# Obtain uncertainty estimates (variance) across trees
y_val_new_pred_all_trees = np.stack([tree.predict(X_val_new_standardized) for tree in rf_model_small.estimators_], axis=0)
y_val_new_pred_variance = np.var(y_val_new_pred_all_trees, axis=0)

# Evaluate performance on the new validation set
mse_val_new_rf = mean_squared_error(y_val_new, y_val_new_pred_rf)
r2_val_new_rf = r2_score(y_val_new, y_val_new_pred_rf)

print(f'Mean Squared Error (New Ensemble): {mse_val_new_rf}')
print(f'R-squared (New Ensemble): {r2_val_new_rf}')

# Visualization with uncertainty estimates
plt.figure(figsize=(10, 6))
plt.scatter(y_val_new, y_val_new_pred_rf, alpha=0.01, label='Actual vs. Predicted (New Ensemble)')

plt.xlabel('Actual Values (New Validation)')
plt.ylabel('Ensemble Predicted Values (New Validation)')
plt.title('New Validation Set Predictions - Ensemble with Uncertainty Estimates')
plt.legend()
plt.show()
```
